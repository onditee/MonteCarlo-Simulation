```python
import numpy as np
import math
n = [(10**3),(10**4),(10**5),(10**6)]
length = 1
radius = 1

for number in n:
    x_1 = np.random.rand(number)*length
    x_2 = np.random.rand(number)*length
    y = np.sqrt((x_1**2) + (x_2**2))
    index = np.where(y <= radius)
    target = y[index]
    pi = 4 *(len(target)/number)
    real_pi = math.pi
    error = (abs(real_pi - pi)/real_pi)*100

    print("n = ",number ,"pi = ", pi, "error = ",error,"%")

```
